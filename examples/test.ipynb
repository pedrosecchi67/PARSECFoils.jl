```julia
using PARSECFoils
```

```julia
using Plots
using LinearAlgebra
```

```julia
using DelimitedFiles
```

```julia
function plotxty!(p, x, t, y; kwargs...)

    plot!(p, x, y .+ t ./ 2; kwargs...)
    plot!(p, x, y .- t ./ 2; kwargs...)

end

plotxty(x, t, y; kwargs...) = let p = plot()
    plotxty!(p, x, t, y; kwargs...)

    p
end
```

```julia
xs = collect(LinRange(0.0, 1.0, 300))

xs, ts, ys = parsec2xty(
    xs
)

plotxty(xs, ts, ys; ylim = (-0.5, 0.5))
```

```julia
parsec2selig(xs)
```

```julia
pts = readdlm("GAW1.dat"; skipstart = 1,)

pars = selig2parsec(xs, pts)

pars
```

```julia
xs, ts, ys = parsec2xty(xs; pars...)

p = plotxty(xs, ts, ys)

plot!(
    p, pts[:, 1], pts[:, 2]
)
```
